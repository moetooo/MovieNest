```python
def stopword(text):
  l=[]
  for i in text.split(' '):
    if i not in stops:
      l.append(i)
  return ' '.join(l)
```

```python
def stem(text) :
  l=[]
  for i in text.split(' '):
    if i not in stops :
      l.append(ps.stem(i))
  return ' '.join(l)
```

```python
import re

def recommend(movie):
    # Clean the movie title (remove non-alphabetic characters)
    movie = re.sub(r'[^a-zA-Z]', '', movie).lower()

    # Find the index of the movie in the dataset
    movie_index = new_df[new_df['clean_title'] == movie].index

    # Check if the movie exists
    if movie_index.empty:  # If movie not found
        print(f"Movie '{movie}' not found in dataset.")
        return

    movie_index = movie_index[0]  # Safely access the first index

    # Get the similarities (you should define or load 'similarity' matrix beforehand)
    difference = similarity[movie_index]

    # Get the top 5 similar movies (excluding the movie itself)
    movies_list = sorted(list(enumerate(difference)), reverse=True, key=lambda x: x[1])[1:6]

    # Print the titles of the top 5 similar movies
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
